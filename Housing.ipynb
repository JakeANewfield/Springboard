```python
# Let's import the pandas, numpy libraries as pd, and np respectively. 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import ssl
ssl._create_default_https_context = ssl._create_unverified_context
```

```python
# First, make a variable called url_LondonHousePrices, and assign it the following link, enclosed in quotation-marks as a string:
# https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls

url_LondonHousePrices = "https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls"

# The dataset we're interested in contains the Average prices of the houses, and is actually on a particular sheet of the Excel file. 
# As a result, we need to specify the sheet name in the read_excel() method.
# Put this data into a variable called properties.  
properties = pd.read_excel(url_LondonHousePrices, sheet_name='Average price', index_col= None)
```

```python
# In[13]:


properties.describe()
properties.head()
```

```python
# In[14]:


#Transposing the df:
properties_T = properties.T
properties_T.head(5)
#Resetting the Index:
df = properties_T.reset_index()
#shortening to 'df'
df.columns = df.iloc[0]

df.head(5)
```

```python
# In[15]:


df = df.rename(columns = {'Unnamed: 0': 'Borough', 'NaT': 'ID'})
df = df.drop(df.index[0])

# Reset the index
df = df.reset_index(drop=True)
df = df.rename(columns={df.columns[1]: 'ID'})
df = df.drop('ID', axis=1)
df.head(10)
```

```python
# **2.4.Transforming the data**
# 
# Remember what Wes McKinney said about tidy data? 
# 
# You might need to **melt** your DataFrame here. 

# In[16]:


df = df.melt(id_vars='Borough', var_name='Date', value_name='Average Price')
df.head(15)
```

```python
# Remember to make sure your column data types are all correct. Average prices, for example, should be floating point numbers... 

# In[17]:


df['Average Price'] = df['Average Price'].astype(float)

df.dtypes
```

```python
# In[18]:


df.count()
#there are roughly 1000 extra rows that need to be deleted. any row with a missing 'average price' should be deleted - right?

df = df.dropna(subset=['Average Price'])

df.count()
#now there are are the same number of valid values in each column.  Was this the right thing to do?

df['Borough'].unique()
```

```python
# **2.6. Visualizing the data**
# 
# To visualize the data, why not subset on a particular London Borough? Maybe do a line plot of Month against Average Price?

# In[19]:


# Extract the columns to compare
column1 = df['Date']
column2 = df['Average Price']

# Create a line plot to compare the columns
plt.plot(column1, label='Date')
plt.plot(column2, label='Average Price')

# Set labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Comparison of Column 1 and Column 2')

# Add legend
plt.legend()

# Display the plot
plt.show()
```

```python
# To limit the number of data points you have, you might want to extract the year from every month value your *Month* column. 
# 
# To this end, you *could* apply a ***lambda function***. Your logic could work as follows:
# 1. look through the `Month` column
# 2. extract the year from each individual value in that column 
# 3. store that corresponding year as separate column. 
# 
# Whether you go ahead with this is up to you. Just so long as you answer our initial brief: which boroughs of London have seen the greatest house price increase, on average, over the past two decades? 

# In[20]:


# Convert 'Month' column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Extract the year and store it in a new column
df['Year'] = df['Date'].dt.year

# Display the updated DataFrame
df
```

```python
# In[26]:


# Group the DataFrame by 'Borough' and calculate the price difference between the earliest and latest years
price_diff = df.groupby('Borough')['Average Price'].agg(lambda x: x.iloc[-1] - x.iloc[0])

# Sort the price differences in descending order
price_diff_sorted = price_diff.sort_values(ascending=False)

# Filter for the top boroughs with the greatest price increase
top_boroughs = price_diff_sorted.head()

#make it readable:
pd.options.display.float_format = '{:.2f}'.format

# Display the top boroughs
top_boroughs


# In[25]:


#what datatype are the results above:
top_boroughs.dtype
```
